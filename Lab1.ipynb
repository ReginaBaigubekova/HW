```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

def p(num):
  for i in documents:
    if i['number'] == num:
      print(f"Владелец документа: {i['name']}")
      break
  else: 
    print("Документ не найден в базе")


def s(num):
  for i in directories:
    if num in directories[i]:
      return i
  return None
      

def l():
  for i in documents:
    print(f"№: {i['number']}, тип: {i['type']}, владелец: {i['name']}, полка хранения: {s(i['number'])}")


def ads(num):
  if num  not in directories.keys():
    directories[num]=[]
    print(f"Полка добавлена. Текущий перечень полок: {list(directories.keys())}")
  else:
    print(f"Такая полка уже существует. Текущий перечень полок: {list(directories.keys())}")


def ds(num):
  if num  not in directories.keys():
    print(f"Такой полки не существует. Текущий перечень полок: {list(directories.keys())}")
  elif directories[num]==[]:
    del directories[num]
    print(f"Полка удалена. Текущий перечень полок: {list(directories.keys())}")
  else:
     print(f"На полке есть документа, удалите их перед удалением полки.\nТекущий перечень полок: {list(directories.keys())}")


def ad(num,typ,owner,shelf):
  if shelf in directories.keys():
    documents.append({'type':typ, 'number':num, 'name':owner})
    directories[shelf].append(num)
    print("Текущий список документов: ")
    print(l())
  else:
     print("Такой полки не существует. Добавьте полку командой ads.")
  

def d(num):
  for i in documents:
    if i['number'] == num:
      documents.remove(i)
      print(f"Документ удален.\nТекущий список документов:")
      print(l())
      break
  else:
    print(f"Документ не найден в базе.\nТекущий список документов:")
    print(l())


def m(num,shelf):
  for i in documents:
    if i['number'] == num and shelf in directories.keys():
      for item in directories.keys():
        if num in directories[item]:
          directories[shelf].append(num)
          directories[item].remove(num)
          print("Документ перемещен.\nТекущий список документов: ")
          print(l())
      break
    elif i['number'] == num and shelf not in directories.keys():
       print(f"Такой полки не существует. Текущий перечень полок: {list(directories.keys())}")
       break
  else:
     print("Документ не найден в базе.\nТекущий список документов: ")
     print(l())
    

while True:
  com = input("Введите команду: ")
  if com == "p":
    num = input("Введите номер документа: ")
    p(num)
  elif com == "s":
    num = input("Введите номер документа:")
    if s(num) != None:
      print(f"Документ хранится на полке: {s(num)}")
    else:
      print("Документ не найден в базе")
   
  elif com == "l":
    l()
  elif com == "ads":
    num = input("Введите номер полки:")
    ads(num)
  elif com == "ds":
    num = input("Введите номер полки:")
    ds(num)
  elif com == "ad":
    num = input("Введите номер документа:")
    typ = input("Введите тип документа:")
    owner = input("Введите владельца документа:")
    shelf = input("Введите номер полки:")
    ad(num,typ,owner,shelf)
  elif com == "d":
    num = input("Введите номер документа:")
    d(num)
  elif com == "m":
    num = input("Введите номер документа:")
    shelf = input("Введите номер полки:")
    m(num,shelf)
  elif com == "q":
    break
```
