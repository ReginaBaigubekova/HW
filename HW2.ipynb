```python
import math
x=input("Какая фигура(круг, треугольник, прямоугольник)? ")
if x=="круг" :
  r=int(input("Введите радиус: "))
  print(f"Площадь круга: {round(math.pi*(r**2) , 2)}")
elif x=="треугольник":
  c = int(input("Введите первую сторону: "))
  d = int(input("Введите вторую сторону: "))
  e = int(input("Введите третью сторону: "))
  p=(c+d+e)/2
  print(f"Площадь треугольника: {math.sqrt(p*(p-c)*(p-d)*(p-e))}")
else:
    k=int(input("Введите длину первой стороны: "));
    l=int(input("Введите длину второй стороны: "))
    print(f"Площадь прямоугольника: {k*l}")
```
