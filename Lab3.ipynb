```python
import json

all_tasks = []
while True:
  command = input('''
        Простой todo:
        1. Добавить задачу.
        2. Вывести список задач.
        3. Выход.
        Укажите число: ''')
  if command == "1":
    name = input("Сформулируйте задачу: ")
    category = input("Добавьте категорию к задаче: ")
    time = input("Добавьте время к задаче: ")
    all_tasks.append({"category": category, "name": name, "time": time})
    file = open("/content/drive/Othercomputers/Ноутбук/УНИВЕР/todo list.json", "w")
    file.write(json.dumps(all_tasks))
    file.close()    
  elif command == "2":
    with open('/content/drive/Othercomputers/Ноутбук/УНИВЕР/todo list.json') as file:
      tasks = json.load(file)
    for item in tasks:
      print(" Задача: ", item["name"], " Категория: ", item["category"], " Дата: ", item["time"])      
  elif command == "3":
    print("Задачи сохранены в файл!")
    break
  else:
    print("Попробуйте еще раз!")
        
```
