```python
import statistics
import requests

temper = requests.get('https://raw.githubusercontent.com/dm-fedorov/python3_intro/master/lesson_10/temper.stat')
res = list(map(float,temper.text.split()))
print(min(res))
print(max(res))
print(statistics.mean(res))
```

```python
try:
    a = int(input("Введите первое число "))
    c = input("Введите оператор(+,-,/) ")
    b = int(input("Введите второе число "))
    if c == '-':
        result = a - b
    elif c == '+':
        result = a + b
    else:
       result = a / b 
    print(result)
except ValueError:
    print("Ошибка преобразования типов")
except ZeroDivisionError:
    print("Ошибка деления на ноль")
```
