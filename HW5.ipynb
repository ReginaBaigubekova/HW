Задача 6

```python
import math

def my_log(lst):
  ans = []
  for item in lst:
    if item > 0:
      ans.append(math.log(item))
    else:
      ans.append(None)
  return ans
  
my_log([1, 3, 2.5, -1, 9, 0, 2.71])
```

Задача 17

```python
import statistics

def mm(lst):
  i_min = lst.index(min(lst))
  i_max = lst.index(max(lst))
  if i_min < i_max:
    return(statistics.mean(lst[i_min+1:i_max]))
  else:
    lst[i_min] = lst[i_max] = (lst[i_min] + lst[i_max])/2
    return lst

mm([5,2,1,6,3,3,9,2,3])
```

```python
import statistics

def find_mean(lst):
  return [item if item else statistics.mean([item for item in lst if item]) for item in lst ]

find_mean([1, 5, 6.3, 6, None, 2.0, -4, None])
```
