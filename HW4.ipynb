Задача 1

```python
import random

num = random.randint(1,10)
your_num = 0
while your_num != num:
  your_num = int(input("Введите число от 1 до 10: "))
  if your_num > num: 
    print("Ваше число больше загаданного")
  elif your_num < num: 
    print("Ваше число меньше загаданного")
  elif your_num == num: 
    print("Вы угадали")
```

Задача 4

```python
import random

def your_password():
  password = []
  for i in range(random.randint(7,10)-1):
    password.append(chr(random.randint(33,126)))
  return  ''.join (password)

def checking(pw):
  return len(pw) >= 8 and pw.lower() != pw and pw.upper() != pw  and any(char.isdigit() for char in pw)

count = 0
while True:
  p = your_password()
  if checking(p) == True:
    print(f"Надежный пароль: {p},   количество попыток: {count}")
    break
  else:
     count+=1
```
