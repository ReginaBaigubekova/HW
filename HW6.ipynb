```python
class Hotel:
    def __init__(self, all_rooms,price):
        self._rooms = {"SGL": {"all_rooms":[0 for _ in range(all_rooms[0])],"price":price[0]},
                       "DBL":{"all_rooms":[0 for _ in range(all_rooms[1])],"price":price[1]},
                       "Junior Suite": {"all_rooms":[0 for _ in range(all_rooms[2])],"price":price[2]},
                       "Suite":{"all_rooms":[0 for _ in range(all_rooms[3])],"price":price[3]}
                       }           

# 1 метод для бронирования номера по уникальному значению в списке
    def occypy(self, typee, room_id):
        if self._rooms[typee]["all_rooms"][room_id]==0:
            self._rooms[typee]["all_rooms"][room_id] = 1 
        else:
            raise RuntimeError("Номер занят")

# 2 метод для освобождения номера по уникальному значению в списке
    def free(self,typee, room_id):
        self._rooms[typee]["all_rooms"][room_id] = 0 

# 3 метод для красивой печати номеров
    def __str__(self):
          a = ''
          for typee in self._rooms:
              for room in range(len(self._rooms[typee]["all_rooms"])):
                  if self._rooms[typee]["all_rooms"][room]==0:
                      a+='Номер '+ typee + " " + str(room + 1) + " свободен\n"
                  else:
                      a+='Номер '+ typee + " " + str(room + 1) + " занят\n"
          return a 
           
# 4 метод, который занимает все свободные номера в отеле
    def all_occypy(self):
        for typee in self._rooms:
              for room in range(len(self._rooms[typee]["all_rooms"])):
                self._rooms[typee]["all_rooms"][room] = 1

# 5 метод, который возвращает долю занятых номеров
    def share_of_occupied_rooms(self):
        all, occupied = 0, 0
        for typee in self._rooms:
          occupied += sum(self._rooms[typee]["all_rooms"])
          all += len(self._rooms[typee]["all_rooms"])
        return f"Доля занятых номеров = {round(occupied*100 / all ,3)}%"

# 6 метод, который освобождает все занятые номера отеля
    def all_free(self):
        for typee in self._rooms:
              for room in range(len(self._rooms[typee]["all_rooms"])):
                self._rooms[typee]["all_rooms"][room] = 0

# 7 метод, который возвращает выручку, исходя из занятости комнат
    def get_revenue(self):
        s=0
        for typee in self._rooms:
            s+=sum(self._rooms[typee]["all_rooms"])*self._rooms[typee]["price"]
        return f"Выручка = {s}"



hotel = Hotel([7,3,1,2],[4000,2000,1000,2500])
print(hotel._rooms) 

hotel.occypy('Suite',0)
hotel.occypy('SGL',4)
hotel.occypy('SGL',5)
hotel.occypy('SGL',6)
hotel.occypy('DBL',1)
print(hotel)
print(hotel.share_of_occupied_rooms())
print(hotel.get_revenue())
print()

hotel.free('Suite',0)
print(hotel)
print()

hotel.all_free()
print(hotel)
print(hotel.share_of_occupied_rooms())
print()

hotel.all_occypy()
print(hotel)
print(hotel.share_of_occupied_rooms())
print()
```
