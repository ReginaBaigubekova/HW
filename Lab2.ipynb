```python
def put(name,surname):
  file=open("/content/drive/Othercomputers/Ноутбук/УНИВЕР/students.txt", "r")
  students = file.readlines()
  students=[i.strip() for i in students]
  file.close()
  if name+" "+surname in students:
    return("Студент уже добавлен")
  else:
    students+=[name+" "+surname]
    students.sort()
    file=open("/content/drive/Othercomputers/Ноутбук/УНИВЕР/students.txt", "w")
    for i in students:
      file.write(i+"\n")
    print("Добавлено")
    file.close()
```

```python
def find(surname,name=None):
  file=open("/content/drive/Othercomputers/Ноутбук/УНИВЕР/students.txt", "r")
  if name:
    for i in file:
      if name+" "+surname in i.strip():
        return("Данный студент находится в группе")
    else:
      return("Данный студент не находится в группе")
  else:
    for i in file:
      if surname in i.strip():
        print(i.strip())
  file.close()
```

```python
def edit(name,surname,new_name=None,new_surname=None):
  file=open("/content/drive/Othercomputers/Ноутбук/УНИВЕР/students.txt", "r")
  students = file.readlines()
  students=[i.strip() for i in students]
  file.close()
  if name+" "+surname not in students:
    return("Данный студент не находится в группе")
  else:
    if new_name==None and new_surname==None:
      print("Изменения не указаны")
    else:
      if new_name==None: new_name=name
      if new_surname==None: new_surname=surname
      delete(surname,name)
      put(new_name,new_surname)
      print("Изменено")
```

```python
def delete(surname,name=None):
  file=open("/content/drive/Othercomputers/Ноутбук/УНИВЕР/students.txt", "r")
  students = file.readlines()
  students=[i.strip() for i in students]
  file.close()
  if name==None:
    find(surname)
    name=input("Выберете имя студента ")
  if name+" "+surname not in students:
    return("Данный студент не находится в группе")
  else:
    students.remove(name+" "+surname)
    file=open("/content/drive/Othercomputers/Ноутбук/УНИВЕР/students.txt", "w")
    for i in students:
      file.write(i+"\n")
    print("Студент удален")
    file.close()
  
```
